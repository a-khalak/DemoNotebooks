```python
import wikipedia as wp
import pandas as pd
```

```python
#Get the html source
html = wp.page("Major Diagnostic Codes").html().encode("UTF-8")
df = pd.read_html(html)[0]
df.head()
```

```python
df.columns = df.iloc[0].values.flatten().tolist()
df = df.drop([0], axis=0)
df.rename(columns = {'MS-DRG[1][2]': 'ms_drg_range', 'MDC' : 'mdc', 
                     'Description' : 'description'}, inplace=True)
```

```python
df['ms_drg_start'] = df.ms_drg_range.apply(lambda x: x.split(' - ')[0])
df['ms_drg_end'] = df.ms_drg_range.apply(lambda x: x.split(' - ')[1])
```
